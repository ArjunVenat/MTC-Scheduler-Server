<h2><u>Formulation</u></h2>


<h4><u>Set Definition</u></h4>

Let $P$ be the set of PLAs, indexed by $i$

Let $G$ be the set of GPLs, indexed by $i$
- Let $G_b$ be the subset of GLAs who have indicated a preference for back-to-back shifts
- Let $G_d$ be the subset of GLAs who have indicated a preference for same-day shifts

Let $T$ be the set of TAs, indexed by $i$
- Let $T_b$ be the subset of TAs who have indicated a preference for back-to-back shifts
- Let $T_d$ be the subset of TAs who have indicated a preference for same-day shifts

Let $D$ be the set of days the MTC is open, indexed by $j$

Let $S$ be the set of shifts the MTC is open on day $j$, indexed by $k$ 
- May change the notation for this, either an adjacency list (the set of shifts depends on the day) or adjacency matrix (there are $|D|$ * $|S|$ total day-shift pairs where in this case, $|S|$ is the maximum number of shifts in a day across all days. The latter seems more elegant and is consistent with the other notation)

<h4><u>Constant Definition</u></h4>

Let $l_{jk}$ be the minimum number of student workers for 
day $j$ shift $k$

Let $u_{jk}$ be the maximum number of student workers for 
day $j$ shift $k$

Let $x_{ijk}$ be the preference indicated by worker $i$ for day $j$ shift $k$, where $x_{ijk} \in \{1, 2, 3, ∞\}$

<h4><u>Variable Definition</u></h4>

Let $a_{ijk}$ be the decision to assign student worker $i$ to day $j$ shift $k$, where $a_{ijk} \in \{0, 1\}$.

Let $a_{ij}$ be the binary assignment vector for all shifts on day $j$ for student worker $i$, where $a_{ij} = ( a_{ij1}, a_{ij2}, ..., a_{ijk} ) $

<h4><u>Mathematical Formulation</u></h4>

$minimize$ $\sum\limits_{i \in P∪G∪T} \sum\limits_{j \in D} \sum\limits_{k \in S} x_{ijk}*a_{ijk}$

$s.t$

$\sum\limits_{j \in D} \sum\limits_{k \in S} a_{ijk} = 1 \;\; \forall i \in P$

$\sum\limits_{j \in D} \sum\limits_{k \in S} a_{ijk} = 2 \;\; \forall i \in G∪T$

$\sum\limits_{i \in P∪G∪T} a_{ijk} \le u_{jk} \;\; \forall j \in D, k \in S$

$\sum\limits_{i \in P∪G∪T} a_{ijk} \ge l_{jk} \;\; \forall j \in D, k \in S$

$\sum\limits_{j \in D} a_{ij} \land (a_{ij} \gg 1) > 0 \;\; \forall i \in G_b∪T_b$

$\sum\limits_{k \in S} a_{ijk} \in \{0, n\} \;\; \forall i \in G_d∪T_d, j \in D$ where $n$ is the number of shifts worked by worker $i$.
