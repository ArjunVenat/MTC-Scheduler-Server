<h2><u>Formulation</u></h2>


<h4><u>Set Definition</u></h4>

Let $P$ be the set of PLAs, indexed by $i$

Let $G$ be the set of GLAs/Graders, indexed by $i$
- Let $G_b$ be the subset of GLAs/Graders who have indicated a preference for back-to-back shifts
- Let $G_d$ be the subset of GLAs/Graders who have indicated a preference for same-day shifts

Let $T$ be the set of TAs, indexed by $i$
- Let $T_b$ be the subset of TAs who have indicated a preference for back-to-back shifts
- Let $T_d$ be the subset of TAs who have indicated a preference for same-day shifts

Let $M$ be the set of student workers who are willing to work more than one shift, indexed by $i$ #naturally $M$ is a superset of $T$
- Let $M_b$ be the subset of TAs who have indicated a preference for back-to-back shifts
- Let $M_d$ be the subset of TAs who have indicated a preference for same-day shifts


Let $D$ be the set of days the MTC is open, indexed by $j$

Let $S$ be the union of shifts the MTC is open across all days, indexed by $k$ 

<h4><u>Constant Definition</u></h4>

Let $l_{jk}$ be the minimum number of student workers for day $j$ shift $k$

Let $u_{jk}$ be the maximum number of student workers for day $j$ shift $k$

Let $c_{i}$ be the maximum number of shifts that worker $i$ is willing to work. 

Let $x_{ijk}$ be the preference indicated by worker $i$ for day $j$ shift $k$, where $x_{ijk} \in \{1, 2, 3, 10000\}$

<h4><u>Variable Definition</u></h4>

Let $a_{ijk}$ be the decision to assign student worker $i$ to day $j$ shift $k$, where $a_{ijk} \in \{0, 1\}$.

Let $a_{ij}$ be the binary assignment vector for all shifts on day $j$ for student worker $i$, where $a_{ij} = ( a_{ij1}, a_{ij2}, ..., a_{ijk} ) $

<h4><u>Mathematical Formulation</u></h4>

$minimize$ $\sum\limits_{i \in P∪G∪T} \sum\limits_{j \in D} \sum\limits_{k \in S} x_{ijk}*a_{ijk}$

$s.t$

$\sum\limits_{j \in D} \sum\limits_{k \in S} a_{ijk} >= 1 \;\; \forall i \in PUG \;\;$ #PLAs and GLAs/Graders are assigned to at least one shift

$\sum\limits_{j \in D} \sum\limits_{k \in S} a_{ijk} >= 2 \;\; \forall i \in T \;\;$ #TAs are assigned to at least two shifts

$\sum\limits_{j \in D} \sum\limits_{k \in S} a_{ijk} <= c_{i} \;\; \forall \in {PUGUT} \;\;$ #Workers will not more shifts than they are willing to

$\sum\limits_{i \in P∪G∪T} a_{ijk} \le u_{jk} \;\; \forall j \in D, k \in S \;\;$ #Upper bound on number of workers on a shift

$\sum\limits_{i \in P∪G∪T} a_{ijk} \ge l_{jk} \;\; \forall j \in D, k \in S \;\;$ #Lower bound on number of workers on a shift

$\sum\limits_{j \in D} a_{ij} \land (a_{ij} \gg 1) > 0 \;\; \forall i \in M_b \;\;$ #Logical and between [worker, day, :] vector and itself shifted by one to the right. The vector evaluates to > 1 if back-to-back is fulfilled and 0 if not, so this will constrain the decision variables to be back to back.

$\sum\limits_{k \in S} a_{ijk} \in \{0, n\} \;\; \forall i \in M_d, j \in D$ where $n$ is the number of shifts worked by worker $i$. #Same day constraint
